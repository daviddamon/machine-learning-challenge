```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("data/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
pd.set_option('display.max_columns', None)
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.info()
```

# Select your features (columns)

```python
data = df.drop("koi_disposition", axis=1)
data.head()
```

```python
# Set features. This will also be used as your x values.
#selected_features = df[['names', 'of', 'selected', 'features', 'here']]
selected_features = df[['koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec', 'koi_time0bk',                        'koi_period','koi_period_err1','koi_period_err2',
                        'koi_time0bk_err1','koi_time0bk_err2', 'koi_impact','koi_impact_err1',
                        'koi_impact_err2', 'koi_duration','koi_duration_err1','koi_duration_err2',
                        'koi_depth','koi_depth_err1','koi_depth_err2', 'koi_prad','koi_prad_err1',
                        'koi_prad_err2', 'koi_teq','koi_insol','koi_insol_err1','koi_insol_err2',
                        'koi_model_snr','koi_steff','koi_steff_err1','koi_steff_err2', 'koi_slogg',
                        'koi_slogg_err1','koi_slogg_err2', 'koi_srad','koi_srad_err1', 'koi_srad_err2',
                        'ra','dec','koi_kepmag']]
selected_features.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
#target = df["koi_disposition"]
#target_names = ["confirmed", "candidate", "false positive"]
```

```python
# assign X (data) and y (target)

X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]

print(X.shape, y.shape)
```

```python
# split the data into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# create a MinMaxScaler model and fit it to the training data

from sklearn.preprocessing import MinMaxScaler, LabelEncoder

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_test_scaled
```

# Train the Model



```python
# support vector machine linear classifier

from sklearn.svm import SVC

model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
predictions = model.predict(X_test)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# create the GridSearchCV model
# create the GridSearch estimator along with a parameter object containing the values to adjust

from sklearn.model_selection import GridSearchCV

param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# fit the model using the grid search estimator. 
# this will take the SVC model and try each combination of parameters

grid.fit(X_train_scaled, y_train)
```

```python
# list the best parameters and score for this dataset

print(grid.best_params_)
print(grid.best_score_)
```

```python
# Train the model with GridSearch
```

```python
grid.score(X_train_scaled, y_train)
```

```python
predictions = grid.predict(X_test_scaled)
print(predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
