```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("../data/exoplanet_data.csv")

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
pd.set_option('display.max_columns', None)

# Drop the null rows
df = df.dropna()
df.head()
```

```python
#df.info()
```

# Select your features (columns)

```python
# set features to be used as x values
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]

selected_features = df
selected_features.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# assign X (data) and y (target)

X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]

print(X.shape, y.shape)
```

```python
# split the data into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# create a MinMaxScaler model and fit it to the training data

from sklearn.preprocessing import MinMaxScaler, LabelEncoder

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_test_scaled
```

# Train the Model



```python
# create a Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train)
rf.score(X_train_scaled, y_train)
```

```python
# automatically calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# sort the features by their importance
sorted(zip(rf.feature_importances_, selected_features), reverse=True)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# create the GridSearchCV model
# create the GridSearch estimator along with a parameter object containing the values to adjust

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 500, 1000],
    'max_depth': [10, 100, 1000]
}
grid = GridSearchCV(rf, param_grid, verbose=3)
```

```python
# fit the model using the GridSearch estimator. 
# this will take the SVC model and try each combination of parameters

grid.fit(X_train_scaled, y_train)
```

```python
# list the best parameters and score for this dataset

print(f"Best Parameters: {grid.best_params_}")

print(f"Best Score: {grid.best_score_}")
```

```python
# Train the model with GridSearch

train_grid_score = grid.score(X_train_scaled, y_train)
test_grid_score = grid.score(X_test_scaled, y_test)

print(f"Training Grid Score: {train_grid_score}")
print(f"Testing Grid Score: {test_grid_score}")
```

```python
predictions = grid.predict(X_test_scaled)
print(predictions)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions))
```

# Save the Model

```python
# save the model

import joblib

filename = 'Random_Forest_model.sav'
joblib.dump(rf, filename)
```
